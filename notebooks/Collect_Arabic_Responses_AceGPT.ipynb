**Load Data**

```python
import pandas as pd

df = pd.read_excel("/home/yasser.attia/Llama/Regional_specific_dataset_for_arab_countries_updated.xlsx") #changes this for  egypt specific data and general data with the file path
```

```python
arabic_questions = df['Questions in Arabic']

count = 0
for i, q in arabic_questions.items():
    if pd.isna(q):
        count += 1
print(count)
model_inputs = list(arabic_questions)
print(len(model_inputs))
```

**Load Model**

```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("FreedomIntelligence/AceGPT-7B-chat")
model = AutoModelForCausalLM.from_pretrained("FreedomIntelligence/AceGPT-7B-chat")
```

```python
import torch

# Check if CUDA is available
if torch.cuda.is_available():
    device_id = 3  # You can change this to the appropriate GPU id if you have multiple GPUs
    device = f"cuda:{device_id}"
else:
    device = "cpu"
model.to(device)
```

**Collect Responses**

```python
responses = []
for i in range(len(model_inputs)):
    batch = model_inputs[i]
    print(batch)
    inputs = tokenizer(batch, return_tensors="pt").to(device)
    outputs = model.generate(**inputs, max_length=1024)
    result = tokenizer.decode(outputs[0], skip_special_tokens=True)
    responses.append(result)
```

**Merge responses to original data**

```python
if len(df.columns) < 11:
    df["Unnamed: 10"] = pd.NA  

df.loc[len(responses), df.columns[10]] = responses

# Save the updated DataFrame back to an Excel file
df.to_excel('Regional_specific_dataset_for_arab_countries_updated.xlsx', index=False)
```
