```python
import matplotlib.pyplot as plt
import seaborn as sns
Res_dir=np.array(Res_dir)
plt.figure(figsize = (7,4))
ax = sns.heatmap(data=Res_dir, annot=True, 
                 annot_kws={"fontsize":12}, fmt=".0f", cmap = sns.cm.rocket_r)
ax.set(xlabel="", ylabel="")

column_labels = ['0','1', '2', '3', '4', '5'] 
ax.set_xticklabels(column_labels[:Res_dir.shape[1]], rotation=0, fontsize=15)

ax.set_yticklabels(['Jais','ACE-GPT','Qwen2','GPT-4o','Llama3'], 
                   rotation=0, fontsize=14)

plt.title('Question Type: Direct attack',fontsize=12)
plt.gcf().savefig("direct_attack.pdf", format='pdf', bbox_inches='tight')

plt.tight_layout()
```
