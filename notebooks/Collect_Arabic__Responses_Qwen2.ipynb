**Load Data**

```python
import pandas as pd

df = pd.read_excel("Regional_specific_dataset_for_arab_countries_updated.xlsx")
print(len(df), df.columns)
```

```python
arabic_questions = df['Questions in Arabic']

count = 0
for i, q in arabic_questions.items():
    if pd.isna(q):
        count += 1
print(count)
model_inputs = list(arabic_questions)
print(len(model_inputs))
```

```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("Qwen/Qwen2-7B-Instruct")
model = AutoModelForCausalLM.from_pretrained("Qwen/Qwen2-7B-Instruct")
```

```python
import torch

# Check if CUDA is available
if torch.cuda.is_available():
    device_id = 3  # You can change this to the appropriate GPU id if you have multiple GPUs
    device = f"cuda:{device_id}"
else:
    device = "cpu"
model.to(device)
```

```python
responses = []
mnm=2200
for i in range(mnm,len(model_inputs)):
    batch = model_inputs[i]
    print(batch)
    inputs = tokenizer(batch, return_tensors="pt").to(device)
    outputs = model.generate(**inputs, max_length=2048)
    result = tokenizer.decode(outputs[0], skip_special_tokens=True)


    #result = model.generate(batch, max_output_length=512, system_prompt = system_prompt)
    responses.append(result)
```
